# 29 — Pole Figure Construction

For texture analysis (preferred orientation, ODF), you need a **pole
figure** — intensity of a specific (h,k,l) ring as a function of the
sample-frame orientation (typically two angles: tilt χ and rotation φ).

`texture.pole_figure.cake_to_pole_figure` regrids the (R, η) cake onto
the (χ, φ) sample frame using the goniometer transform.
`write_popla_pol` writes the result in POPLA `.pol` format consumed by
MAUD / MTEX / popLA.

## Workflow

1. Integrate every frame in your texture scan with `integrate_subpixel`
   or `integrate_polygon`.
2. Stack the cakes into shape `(n_frames, n_eta, n_R)`.
3. For each ring of interest, call `cake_to_pole_figure` with the
   per-frame goniometer angles.
4. Write `.pol` files for MAUD / MTEX consumption.

For a complete texture pipeline including ODF inversion, hand the .pol
files to your favourite analyser — MIDAS doesn't (yet) compute the ODF.
