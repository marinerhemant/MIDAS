# 27 — DAC Gasket Masking

Diamond Anvil Cell (DAC) experiments produce diffraction images with
**huge sectors blocked** by the gasket walls. Naively azimuthally
integrating across blocked sectors washes out the signal and gives
wrong intensities.

`dac.gasket_mask.build_gasket_mask` produces a 2-D mask that excludes
the blocked η ranges. `eta_coverage_per_ring` reports how much of each
ring is actually visible — useful as a quality gate.

````python
import os; os.environ.setdefault('KMP_DUPLICATE_LIB_OK', 'TRUE')
import math, numpy as np, torch, matplotlib.pyplot as plt
````

````python
from midas_integrate_v2.dac import build_gasket_mask, eta_coverage_per_ring

# Typical Princeton/Bassett DAC: ~120° gasket opening symmetric about eta=0.
# Equivalent: block eta in (60°, 180°) and (-180°, -60°).
mask_blocked_intervals = [(60.0, 180.0), (-180.0, -60.0)]

# build_gasket_mask takes the spec + intervals and produces a (NrPixelsZ, NrPixelsY) bool mask.
# True = blocked (exclude from integration).
print('Use build_gasket_mask(spec, blocked_eta_intervals=mask_blocked_intervals)')
````

## Eta coverage gate

`eta_coverage_per_ring(spec, mask)` returns the fraction of each ring's
eta circumference that's actually visible. Rings with coverage below
some threshold (say 0.4) should be excluded from refinement — they're
biased by the asymmetric eta sampling.

