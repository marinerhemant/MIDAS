# 31 — Multi-Detector Streaming

For 1ID-style multi-detector setups (multiple GE / Pilatus modules
covering different Q ranges) you want to **integrate each detector
independently** to its native Q range, then **stitch** the resulting
1-D profiles into a single combined Q-grid.

`streaming.multi_detector.integrate_multi_detector` does this.

```python
import os; os.environ.setdefault('KMP_DUPLICATE_LIB_OK', 'TRUE')
import math, numpy as np, torch, matplotlib.pyplot as plt
```

```python
from midas_integrate_v2.streaming import integrate_multi_detector

# Each detector has its own IntegrationSpec; integrate_multi_detector
# accepts a list of (spec, image) pairs plus a common Q grid.
print('integrate_multi_detector(specs_and_images, Q_grid_invA) -> 1-D profile on Q_grid')
print('Internally:')
print('  for each (spec, img):  integrate to Q via build_q_bin_edges_in_R')
print('  interpolate each profile onto the common Q_grid')
print('  stitch with overlap weighting')
```

## Typical setup

| detector | Lsd | Q range |
|---|---|---|
| GE3 close  | 250 mm | 0–10 Å⁻¹  |
| GE3 mid    | 600 mm | 0–4 Å⁻¹  |
| GE3 far    | 1500 mm | 0–1.6 Å⁻¹ |

`integrate_multi_detector` returns a single profile spanning the union
of all detector Q ranges, with overlap-weighted intensities in shared
regions.
