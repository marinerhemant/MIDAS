# 30 — GISAXS Regrid + Inelastic Regrid

Two non-powder modes that live in `midas_integrate_v2`:

* **GISAXS** (grazing-incidence SAXS): pixels live in (qy, qz), not (R, η).
  `grazing.gisaxs` provides `pixel_to_qy_qz` + `remap_to_qy_qz_grid`.
* **Inelastic** (IXS, RIXS): energy-resolved scattering needs a (Q, E)
  regrid. `inelastic.regroup_eta_R_E_to_Q_E` handles it.

```python
import os; os.environ.setdefault('KMP_DUPLICATE_LIB_OK', 'TRUE')
import math, numpy as np, torch, matplotlib.pyplot as plt
```

```python
from midas_integrate_v2.grazing import pixel_to_qy_qz, remap_to_qy_qz_grid
from midas_integrate_v2.inelastic import regroup_eta_R_E_to_Q_E

# GISAXS: produce per-pixel (qy, qz)
print('pixel_to_qy_qz(spec, alpha_i=...) -> (qy, qz) per pixel')
print('remap_to_qy_qz_grid(img, spec, alpha_i=..., grid=...) -> 2-D qy-qz map')

# Inelastic: regrid eta-R-E -> Q-E for RIXS-style 2-D maps
print('regroup_eta_R_E_to_Q_E(eta_R_E_cube, spec, energies=...) -> (Q, E, I) 2-D')
```

These are minimal wrappers — they share the same `IntegrationSpec`
calibration as the standard radial integrator. If you've calibrated for
powder you can run GISAXS on the same image with no extra setup beyond
the grazing-incidence angle `alpha_i`.
